# Method-2 np.where()

```python
##################### read the data #############################

file_path="C:\\Users\\sony\\ DATA SCIENCE  NARESH IT\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
visa_df
```

- np.where() is applicable for binary condition

- which means it is applicable only for two lables

- np.where(condition,True_value,False_value)

- For example case_status has two lables

- condition: col_values=='Certified'
    
- True value: Replace all certified values with 0
    
- False value: Replace all denied values with 1    
    

```python
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
```

## LableEncoder

- LableEncoder is a method from sklearn

- under sklearn we have sub modules

- one of the submodule: preprocessing
    
- Any sklearn packages we have only 3 step  

- step-1: Read the package
    
- step-2: save the package
    
- step-3: Applyfit transform    
    

```python
###################### Read data #################################

file_path="C:\\Users\\sony\\ DATA SCIENCE  NARESH IT\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
print(visa_df[['continent','case_status']].head(10))

########################## LableEncoder ###########################

# Step-1:
from sklearn.preprocessing import LabelEncoder

#step-2:
le=LabelEncoder()

#step-3:
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df['continent']=le.fit_transform(visa_df['continent'])
print(visa_df[['continent','case_status']].head(10))
```

```python
print(visa_df['continent'][:5])
le.inverse_transform(visa_df['continent']) # values will becomes back
```
