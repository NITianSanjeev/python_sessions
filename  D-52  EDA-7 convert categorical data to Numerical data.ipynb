## One hot encoder

- one hot encoder means at a time only one will be ON(1/True),other are OFF(0/False)

- Suppose case status has two unique lables

- certified and denied

- In one hot encoder it will create two new columns

- like below

|case_status  |  certified |  denied
|-------------|------------|--------
|certified    |    1       |     0
|-------------|------------|--------
|denied       |    0       |    1

### Advantage

- One hot encoder new columns are orthogonal each other

- orthogonality means 90 degree phase shift

- 90 degree phase shift means perpendicular each other

- perpendicular means independent each other 

- Independent means no relation

- It is very important property that is to avoid relation between varaibles

### Draw back

- Assume that you have 100 unique labels for a column

- So it will create 100 new columns

- To process 100 columns we require more memory,more time,more hardware

- This is called **curse of Dimensionality**
