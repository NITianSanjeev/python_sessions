```python
img = cv2.imread('C:\\Users\\sony\\ DATA SCIENCE  NARESH IT\\invoice-sample.jpg')
h, w, c = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitline():
    b = b.split(' ')
    # I 128 983 163 1063
    img= cv2.rectangle(img,
                      (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])),
                      (0,255,0),
                      1)
    
    cv2.imshow('img', img)
    cv2.waitkey(0)
```
